Analysing hate crime in Europe to identify the hate bias and the targeted groups 

```python
#Importing the necessary libraries here
import os
import pandas as pd
import numpy as np
from pprint import pprint

%matplotlib inline
```

```python
#Loading the dataset 'cases-and-rulings-full.csv' into a dataframe
dataframe = pd.read_csv('cases-and-rulings-full.csv' , encoding = "utf8")

dataframe
```

```python
#Checking for null values
dataframe.isnull().sum()
```

```python
#Exploring the columns
dataframe.columns
```

```python
#Dropping the columns that are not useful to train our machine learning models

dataframe.drop(['Case reference', 'URL','Decision / ruling / judgment Date', 'Court / Body type', 'Court / Body', 'KEY FACTS',
       'Main reasoning / argumentation','Key issues (concepts, interpretations) clarified by the case',
       'Results (sanctions, outcome) and key consequences or implications of the case',
       'Key quotation in original language and translated into English  with reference details'], axis = 1, inplace = True)
dataframe

```

```python
#Exploring the list of countries
dataframe.Country.unique()
```

```python
from sklearn import preprocessing 

# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 

# Encoding the labels in column. 
dataframe['country_encoded']= label_encoder.fit_transform(dataframe['Country']) 
dataframe['crimes_encoded'] = label_encoder.fit_transform(dataframe['Crime type(s)'])
dataframe['bias groups']= label_encoder.fit_transform(dataframe['GROUPS']) 
dataframe['hate_bias']= label_encoder.fit_transform(dataframe['HATE BIAS']) 

dataframe.head(10)
```

```python
#Before we instantiate our ML model, we divide the dataset into two smaller training and testing datasets.
#We also set up our Features (X) and Labels (y) where X is 'country_encoded', 'crimes_encoded', and 'bias groups'.
#y is 'HATE BIAS'

from sklearn.model_selection import train_test_split

# test size represents the proportion of training and testing data split. 
X = dataframe[['country_encoded', 'crimes_encoded', 'bias groups']]
Y = dataframe['HATE BIAS']

#Random_state is set to 1 to define the "randomness" of data randomisation
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
# Importing  MultinomialNB from sklearn
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics


# Creating an instance of the model
NB_model = MultinomialNB()

# Fitting the model to training data
NB_model.fit(X_train, Y_train)
```

```python
# Predicting answers to data from the X_text dataset
NB_model_predictions = NB_model.predict(X_test)

# Showing the results in a confusion matrix
print(metrics.confusion_matrix(Y_test,NB_model_predictions))
```

```python
#Printing the classification report and the accuracy
print(metrics.classification_report(Y_test,NB_model_predictions))
```

```python
print(metrics.accuracy_score(Y_test,NB_model_predictions))
```

```python
#Importing the Random Forest Classifier model
from sklearn.ensemble import RandomForestClassifier

# Creating an instance of the model
rf_model = RandomForestClassifier()

# Fitting the model on training data
rf_model.fit(X_train, Y_train)
```

```python
# Predicting the answers to data from the X_text dataset
rf_model_predictions = rf_model.predict(X_test)

# Printing the confusion matrix
print(metrics.confusion_matrix(Y_test,rf_model_predictions))
```

```python
#Printing the classification report and the accuracy
print(metrics.classification_report(Y_test,rf_model_predictions))
```

```python
print(metrics.accuracy_score(Y_test,rf_model_predictions))
```

```python
# Importing the Support Vector Classification model (SVC)
from sklearn.svm import SVC

# The gamma value is set to "auto" to avoid returning error. We fit the SVC model on our data. The confusion matrix is printed.
svc_model = SVC(gamma="auto")
svc_model.fit(X_train, Y_train)

svc_model_predictions = svc_model.predict(X_test)

print(metrics.confusion_matrix(Y_test, svc_model_predictions))
```

```python
print(metrics.classification_report(Y_test,svc_model_predictions))
```

```python
print(metrics.accuracy_score(Y_test,svc_model_predictions))
```

```python
#We import LogisticRegression from sklearn
from sklearn.linear_model import LogisticRegression
#We build the model with the L-BFGS option.
#Once the model is built, the training data is then provided to the model.

log_reg_model = LogisticRegression(solver = "lbfgs")

log_reg_model.fit(X_train, Y_train)
```

```python
#We test the accuracy of the model using the test data.
#We import the sklearn.metrics module which includes score functions, performance metrics and pairwise metrics and distance computations. 
#We create a predictions set with some test data. This is unseen data. 

from  sklearn import metrics

# Creating a prediction set here
# Unseen contents of X_test
log_reg_model_predictions = log_reg_model.predict(X_test)
```

```python
# This is the predicted output from the model
log_reg_model_predictions
```

```python
#Printing the confusion matrix, classification report and accuracy
print(metrics.confusion_matrix(Y_test,log_reg_model_predictions))
```

```python
print(metrics.classification_report(Y_test,log_reg_model_predictions))
```

```python
print(metrics.accuracy_score(Y_test,log_reg_model_predictions))
```

```python
#Before we instantiate our ML model, we divide the dataset into two smaller training and testing datasets.
#We also set up our Features (X) and Labels (y) where X is 'country_encoded', 'crimes_encoded', 'hate_bias'.
#y is 'GROUPS'
from sklearn.model_selection import train_test_split

# test size represents the proportion of training and testing data split. 
X = dataframe[['country_encoded', 'crimes_encoded', 'hate_bias']]
y = dataframe['GROUPS']

# Random_state sets "randomness" of data randomisation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
#Importing the random forest classification model
from sklearn.ensemble import RandomForestClassifier

# Create an instance of the model
RF_model = RandomForestClassifier()

# Fit model to training data
RF_model.fit(X_train, y_train)
```

```python
# Predicting answers to data from the X_text dataset
RF_model_predictions = RF_model.predict(X_test)

# Showing results in a confusion matrix
print(metrics.confusion_matrix(y_test,RF_model_predictions))
```

```python
#Printing the classification report and the accuracy
print(metrics.classification_report(y_test,RF_model_predictions))
```

```python
print(metrics.accuracy_score(y_test,RF_model_predictions))
```
